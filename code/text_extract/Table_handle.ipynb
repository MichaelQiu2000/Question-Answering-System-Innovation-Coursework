## 处理第一个理财要素表

```python
df = clean_tables['招商银行零售青葵系列一年半定开5号理财计划产品说明书.xml'][0]
df
```

```python
def clear_blank(string):
    try:
        result = ''.join(string.split(' '))
    except:
        result = string
        
    return result


def handle_subscribe_limit(string):
    string = clear_blank(string)
    # 如果分A、B、C份额，则只提取A份额
    if 'A份额' in string:
        pattern = re.compile(r'^[A-Za-z0-9][\u4e00-\u9fa5]+[:：][\u4e00-\u9fa5_a-zA-Z0-9\/]+[;；。，]')
        string = pattern.search(string).group()
        string = string[4:]
    # 提取上限
    pattern = re.compile(r'[0-9]+[\u4e00-\u9fa5]+[元]')
    string = pattern.search(string).group()
    
    return string


def handle_due_date(string):
    string = clear_blank(string)
    # 提取日期
    pattern = re.compile(r'[0-9]+年+[0-9]+月[0-9]+日')
    string = pattern.search(string).group()
    
    return string
    

def handle_return_rate(string):
    string = clear_blank(string)
    # 如果分A、B、C份额，则只提取A份额
    if 'A份额' in string:
        pattern = re.compile(r'^[A-Za-z0-9][\u4e00-\u9fa5]+[:：][\u4e00-\u9fa5_a-zA-Z0-9\/%.-]+[;；。，]')
        string = pattern.search(string).group()
        string = string[4:]
    pattern = re.compile(r'[0-9.]+%')
    string = pattern.findall(string)
    string = '-'.join(string)
    
    return string


def handle_subscribe_amount(string):
    string = clear_blank(string)
    # 如果分A、B、C份额，则只提取A份额
    if 'A份额' in string:
        pattern = re.compile(r'^[A-Za-z0-9][\u4e00-\u9fa5]+[:：][\u4e00-\u9fa5_a-zA-Z0-9\/%.-]+[;；。，]')
        string = pattern.search(string).group()
        string = string[4:]
    # 提取上限
    pattern = re.compile(r'[上][限][0-9]+[\u4e00-\u9fa5]+')
    up = pattern.search(string)
    if up is None:
        up = ''
    else:
        up = up.group()
    # 提取下限
    pattern = re.compile(r'[下][限][0-9]+[\u4e00-\u9fa5]+')
    down = pattern.search(string)
    if down is None:
        down = ''
    else:
        down = down.group()
    string = '-'.join([down[2:], up[2:]])
    
    return string


def handle_subscribe_init(string):
    string = clear_blank(string)
    # 如果分A、B、C份额，则只提取A份额
    if 'A份额' in string:
        pattern = re.compile(r'[A份额][\u4e00-\u9fa5]+[0-9]+[\u4e00-\u9fa5]+')
        string = pattern.search(string).group()
        string = string[string.index('A份额')+3:]
    else:
        pattern = re.compile(r'[最低][\u4e00-\u9fa5]+[0-9]+[\u4e00-\u9fa5]*[份]')
        string = pattern.search(string).group()
    pattern = re.compile(r'[0-9]+[\u4e00-\u9fa5]*份')
    string = pattern.findall(string)[0]
    return string


def handle_subscribe_schedule(string):
    string = clear_blank(string)
    pattern = re.compile(r'[0-9年月日:：]+')
    string = pattern.findall(string)
    string = '-'.join(string)
    return string
     
    
def extract_info_from_table_1(table):
    table = table.copy()
    table.columns=['element','describes']
    info = {}
    
    table['element'] = table['element'].map(lambda x: clear_blank(x))
    
    # 从名称中提取产品名称和产品代码
    element = clear_blank(table[table['element']=='名称'].describes.values[0])
    info['产品名称'] = element[:element.find('（')]
    info['产品代码'] = element[element.find('：')+1:-1]
    table.drop(table[table['element']=='名称'].index,axis=0,inplace=True)
    
    # 从全国银行业理财信息登记系统登记编号中提取产品登记编号
    element = clear_blank(table[table['element']=='全国银行业理财信息登记系统登记编号'].describes.values[0])
    pattern = re.compile(r'^[\u4e00-\u9fa5]*[：][A-Za-z0-9]*')
    string = pattern.search(element).group()
    info['产品登记编码'] = string[string.find('：')+1:]
    table.drop(table[table['element']=='全国银行业理财信息登记系统登记编号'].index,axis=0,inplace=True)
    
    # 从本金及理财收益（如有，下同）中提取本机及理财收益的说明
    info['本金及理财收益'] = clear_blank(table[table['element']=='本金及理财收益（如有，下同）'].describes.values[0])
    table.drop(table[table['element']=='本金及理财收益（如有，下同）'].index,axis=0,inplace=True)
    
    # 将巨额赎回与大额赎回统一为 巨额赎回
    for i in ['巨额赎回','大额赎回']:
        if i in table['element'].values:
            info['巨额赎回'] = clear_blank(table[table['element']==i].describes.values[0])
            table.drop(table[table['element']==i].index,axis=0,inplace=True)
            
    # 将单笔认购/申购上限、认购/申购上限、单笔认购上限统一为 认购/申购上限
    for i in ['单笔认购/申购上限','认购/申购上限','单笔认购上限']:
        if i in table['element'].values:
            info['认购/申购上限'] = clear_blank(table[table['element']==i].describes.values[0])
            table.drop(table[table['element']==i].index,axis=0,inplace=True)
            
    # 将理财计划预计到期日、理财计划到期日统一为 理财计划到期日
    for i in ['理财计划预计到期日','理财计划到期日']:
        if i in table['element'].values:
            info['理财计划到期日'] = clear_blank(table[table['element']==i].describes.values[0])
            table.drop(table[table['element']==i].index,axis=0,inplace=True)
            
    # 将认购、申购及赎回方式和认购方式统一为 认购、申购及赎回方式
    for i in ['认购、申购及赎回方式','认购方式']:
        if i in table['element'].values:
            info['认购、申购及赎回方式'] = clear_blank(table[table['element']==i].describes.values[0])
            table.drop(table[table['element']==i].index,axis=0,inplace=True)
    
    # 将以下字段删除
    del_element = ['理财计划终止日','理财计划终止','理财计划利益分配','信息披露','发行对象',\
                   '理财产品实际到期日','理财计划实际到期日']
    table.drop(table[table['element'].isin(del_element)].index,axis=0,inplace=True)
    
    for i in table.index:
        if table.loc[i,'element']:
            info[table.loc[i,'element']] = table.loc[i,'describes']
    
    # 处理认购/申购上限
    info['认购/申购上限'] = handle_subscribe_limit(info['认购/申购上限'])
    # 处理理财计划到期日
    info['理财计划到期日'] = handle_due_date(info['理财计划到期日'])
    # 处理业绩比较基准
    info['业绩比较基准'] = handle_return_rate(info['业绩比较基准'])
    # 处理发行规模
    info['发行规模'] = handle_subscribe_amount(info['发行规模'])
    # 处理认购起点
    info['认购起点'] = handle_subscribe_init(info['认购起点'])
    # 处理认购期
    info['认购期'] = handle_subscribe_schedule(info['认购期'])
    
    
    return info
```

```python
df_1 = pd.DataFrame([extract_info_from_table_1(df[0]) for _,df in clean_tables.items()])
```

```python
df_1 = df_1.applymap(lambda x: clear_blank(x))
```

## 处理第二个投资范围表

```python
df = clean_tables['招商银行青葵系列两年定开005号理财计划产品说明书.xml'][1]
df
```

```python
def extract_info_from_table_2(table):
    table = table.copy()
    table.columns = table.iloc[0,:]
    table = table.iloc[1:,:]
    table = table.applymap(lambda x: clear_blank(x))
    table = ''.join(table.to_html(index=False).split('\n')).replace('class', 'cellspacing=\"0\" class')
          
    return table
```

```python
result = [extract_info_from_table_2(df[1]) for _,df in clean_tables.items()]
df_2 = pd.DataFrame(result,columns=['投资策略'])
df_2.columns
```

## 合并成为最终的理财产品属性表格

```python
df = pd.merge(df_1,df_2,left_index=True,right_index=True)
```
